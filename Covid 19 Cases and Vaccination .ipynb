```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1 = 'Indonesia'
selected_country_2 = 'Indonesia'

df1 = pd.read_csv('worldometer_coronavirus_daily_data.csv', header=0)
df1 = df1[df1['country'] == selected_country_1]
df2 = pd.read_csv('country_vaccinations.csv', header=0)
df2 = df2[df2['country'] == selected_country_2]

df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])

print(df1.columns)
print(df2.columns)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plt.plot(df1.index, df1['active_cases'], label = 'cases')
plt.grid(axis='y')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle('Kasus Harian dan Vaksinasi Harian di Indonesia', fontsize=12, fontweight='bold')
ax.set_title('Jumlah Kasus Harian dan Vaksin', fontsize=10)
ax.set_ylabel('Jumlah Kasus')
ax.set_xlabel('')
plt.plot(df1.index, df1['daily_new_cases'], label = 'Cases')
ax1 = plt.twinx()
ax1.plot(df2.index, df2['daily_vaccinations'], 'r', label='Vaccinations')
ax.text(1, -0.3, 'Sumber data: covid19-global-dataset', color='blue', ha='right', transform=ax.transAxes)
ax1.legend()
ax.legend()
plt.grid(axis='y')
ax.plot()
```
